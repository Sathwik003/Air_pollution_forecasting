##Importing the modules

```python
#Importing libraries and loading dataset
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

def parse(x):
    return datetime.strptime(x, '%Y %m %d %H')
                          
orginal_col_names=["No", "year","month", "day", "hour", "pm2.5", "DEWP","TEMP", "PRES", "cbwd", "Iws", "Is", "Ir"]
new_col_names = ['pollution', 'dew', 'temp', 'pressure', 'w_dir', 'w_speed', 'snow', 'rain']   
                            
df = pd.read_csv('/content/drive/MyDrive/Air_Pollution_data_2010-2014.csv',index_col=0, date_parser=parse,parse_dates=[['year', 'month', 'day', 'hour']])  
```

```python
df.head()
```

```python
df.describe()
```

```python
df1 = df.drop('No', axis=1)
df1.head()
```

```python
df1.columns = new_col_names
# df1.index.name = "year_month_day_hour"
df1.head()
```

```python
df1['pollution'].fillna(0, inplace=True)
df1.head(5)
```

```python
df1 = df1[24:] # drop the first day
```

```python
df1.head()
```

```python
df1 = df1.reset_index() 
```

```python
df1.info()
df1.rename(columns={"index":"Date"},inplace=True)
df2 = df1
```

```python
df2.columns.tolist()
```

```python
df3 = df2.resample('D', on='year_month_day_hour').mean()
df3.head()
```

```python
df4 = df3.rename(columns={"year_month_day_hour":"new_date"})
df4.head()
```
