##Importing the modules

```python
#Importing libraries and loading dataset
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

def parse(x):
    return datetime.strptime(x, '%Y %m %d %H')
                          
orginal_col_names=["No", "year","month", "day", "hour", "pm2.5", "DEWP","TEMP", "PRES", "cbwd", "Iws", "Is", "Ir"]
new_col_names = ['pollution', 'dew', 'temp', 'pressure', 'w_dir', 'w_speed', 'snow', 'rain']   
                            
df = pd.read_csv('/content/drive/MyDrive/DMBI Project/Air_Pollution_data_2010-2014.csv',index_col=0, date_parser=parse,parse_dates=[['year', 'month', 'day', 'hour']])  
```

```python
df1 = df.drop('No', axis=1)
df1.head()
```

```python
df1.columns = new_col_names
df1.index.name = "Date_Time"
df1.head()
```

```python
df1.isnull().sum()
```

```python
df2=df1.interpolate(method ='linear', limit_direction ='backward')
df2
```

```python
df2.isnull().sum()
```

```python
df2 = df2.reset_index() 
```

```python
df3 = df2.resample('D', on='Date_Time').mean()
df3.head()
```
